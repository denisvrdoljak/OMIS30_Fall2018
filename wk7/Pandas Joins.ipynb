```python
# Standard Python line to import pandas library
# It's imported as 'pd' for less typing later!

import pandas as pd
import matplotlib.pyplot as plt

# Setting the options to view charts in the notebook
%matplotlib inline

# setting the options to display the whole dataframe
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
# Reading in the file
# Pandas has many built in readers for different file types

df = pd.read_csv("Demographic_Statistics_By_Zip_Code.csv")
```

```python
df.sort_values(['PERCENT RECEIVES PUBLIC ASSISTANCE'], ascending = False)
```

```python
df['US_CITIZEN_%'] = df['PERCENT US CITIZEN'] * 100
df.head()
```

```python
df['PERCENT US CITIZEN'].hist(bins = 25)
```

```python
ax = df['PERCENT US CITIZEN'].hist(bins = 25)
ax.set(xlabel='Percent US Citizen', ylabel='Total Count', title = "Histogram of US Citizenship Percentage")
plt.plot()
```

```python
ax = df['PERCENT US CITIZEN'].hist(bins = 25, color = 'r', alpha = 0.5)
ax.set(xlabel='Percent US Citizen', ylabel='Total Count', title = "Histogram of US Citizenship Percentage")
plt.plot()
```

```python
ts = pd.read_csv("time_series.csv")
```

```python
ts.head()
```

```python
ts.info()
```

```python
ts.Month = pd.to_datetime(ts.Month, format="%Y-%m")
```

```python
ts.info()
```

```python
ts.head()
```

```python
ax = ts.plot(x="Month", y="Snow")
ax.set(xlabel='Month', ylabel='Days', title="# of Snow Days per Month")
plt.plot()
```

```python
ax = ts.plot(x="Month", y=["Snow", "Rain","Sun"])
ax.set(xlabel='Month', ylabel='Days', title="# of Snow Days per Month")
plt.plot()
```

```python
ax = ts.plot(x="Month", y=["Snow", "Rain","Sun"], xticks=range(0, len(ts)))
ax.set(xlabel='Month', ylabel='Days', title="# of Snow Days per Month")
plt.plot()
```
